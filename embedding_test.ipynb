```python
from google.colab import files
import io
from sentence_transformers import SentenceTransformer
import fitz  # PyMuPDF

model = SentenceTransformer("all-MiniLM-L6-v2")

def extract_text_from_pdf(file_bytes: bytes) -> str:
    doc = fitz.open("pdf", file_bytes)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

def generate_embedding(text: str):
    return model.encode([text])[0].tolist()


uploaded = files.upload()

for name, data in uploaded.items():
  pdf_bytes = io.BytesIO(data).read()
  text = extract_text_from_pdf(pdf_bytes)
  embedding = generate_embedding(text)
  print(f"Text extracted from {name}: {text[:200]}...") # Print first 200 characters
  print(f"Embedding generated for {name}: {embedding[:10]}...") # Print first 10 elements of embedding
```

```python
from google.colab import files
import io
from sentence_transformers import SentenceTransformer
import fitz  # PyMuPDF

model = SentenceTransformer("all-MiniLM-L6-v2")

def extract_text_from_pdf(file_bytes: bytes) -> str:
    doc = fitz.open("pdf", file_bytes)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

def generate_embedding(text: str):
    return model.encode([text])[0].tolist()


uploaded = files.upload()

for name, data in uploaded.items():
  pdf_bytes = io.BytesIO(data).read()
  text = extract_text_from_pdf(pdf_bytes)
  embedding2 = generate_embedding(text)
  print(f"Text extracted from {name}: {text[:200]}...") # Print first 200 characters
  print(f"Embedding generated for {name}: {embedding[:10]}...") # Print first 10 elements of embedding
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Assuming 'embedding' is the embedding for the applicant (from the first upload)
# and 'embedding2' is the embedding for the requirements (from the second upload)

# Reshape the embeddings to be 2D arrays as required by cosine_similarity
embedding_applicant = np.array(embedding).reshape(1, -1)
embedding_requirements = np.array(embedding2).reshape(1, -1)

# Calculate cosine similarity
similarity_score = cosine_similarity(embedding_applicant, embedding_requirements)[0][0]

print(f"Cosine similarity between applicant and requirements: {similarity_score}")
```
